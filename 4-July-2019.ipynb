```python
# Linear graph

import matplotlib.pyplot as plt

plt.title('Basic Linear graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time
distance=list(range(2,11,2))
time=list(range(1,6))
#d = 2 * t

plt.title('Distance vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'ro')
#plt.plot(time,distance,'r+')
plt.grid()
plt.show()


```

```python
# v = a * t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]

plt.plot(t,v)
plt.scatter(t,v)
plt.grid()
plt.show()
```

```python
n=list(range(10))
sqn=[i**2 for i in n]
plt.plot(n,sqn)
plt.scatter(n,sqn)
plt.grid()
plt.show()
```

```python
# Voltage - Time
import numpy as np
#t=[i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
v= [1+ np.sin(2*np.pi*i) for i in t]

plt.title('Time vs Voltage Graph')
plt.plot(t,v)
plt.scatter(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Image/Voltage-Time.png')
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Datafiles/income.csv')
years=df.columns[2:]

incomeArakansas=df.values[3,2:]

plt.plot(years,incomeArakansas)
plt.scatter(years,incomeArakansas)
plt.title('Income of Akansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid()
plt.show()
```

```python
# Income of all states from 2005 to 2013
df
incomeallstates=[]

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
    
plt.plot(years,incomeallstates)
plt.scatter(years,incomeallstates)
plt.title('All States Income')
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid()
plt.show()
```
